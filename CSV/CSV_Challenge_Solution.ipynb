### Challenge One

According to the CDC data from 12/28/2016, which three states or territories have the most travel acquired cases? Which state or territory has the most total cases?

```python
### TODO
import csv

with open('cdc-state-case-counts-2016-12-28.csv') as file:
    fileReader = csv.reader(file)
    # skips header
    next(file)
    
    states = [(0, '')]
    for row in fileReader:
        highestCount = states[0][0]
        if int(row[1]) >= highestCount:
            if len(states) < 3:
                states.append((int(row[1]), row[0]))
            else:
                states[0] = (int(row[1]), row[0])
        states = sorted(states)
    
with open('cdc-state-case-counts-2016-12-28.csv') as file:
    fileReader = csv.reader(file)
    # skips header
    next(file)

    state = ''
    count = 0
    for row in fileReader:
        total = int(row[1]) + int(row[2])
        if total >= count:
            state = row[0]
            count = total

            
print(states)
print(state)
```

### Challenge Two

According to the CDC data from 1/10/2018, which states or territories have absolutely no symptomatic cases or presumptive viremic blood donors?

```python
### TODO
import csv

with open('cdc-state-case-counts-2018-01-10.csv') as file:
    fileReader = csv.reader(file)
    # skips header
    next(file)
    
    states = []
    for row in fileReader:
        if row[1] == '0' and row[2] == '0':
            states.append(row[0])

print(states)
```

### Challenge Three

According to Colombia's data from 02/13/2016, how many total cases are there in the country?

```python
### TODO
import csv

with open('colombia-municipal-2016-02-13.csv') as file:
    fileReader = csv.reader(file)
    # skips header
    next(file)
    
    total = 0
    for row in fileReader:
        total += int(row[-1]) # index -1 gives you the last in the list

print(total)
```
