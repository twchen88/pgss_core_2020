Have you heard of the Caesar Cipher? It is a really similar idea to what we just did, but we are going to put some restrictions on it.
You are guaranteed that message is a string, and that you will not shift more than 25 characters in either direction. Capital letters should stay capital and lowercase letters should stay lowercase, and non-letter characters should not be changed. Note that "Z" wraps around to "A". You will write a decoder for the code provided below:

```python
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"


def caesarCipher(message, shiftNum):
    res = ''
    for c in message:
        index = alphabet.find(c)
        if index == -1:
            res += c
        elif index < 26:
            newChar = alphabet[(index + shiftNum) % 26]
            res += newChar
        else:
            newChar = alphabet[(index + shiftNum) % 26 + 26]
            res += newChar
    return res


assert(caesarCipher("We Attack At Dawn", 1) == "Xf Buubdl Bu Ebxo")
assert(caesarCipher("1234", 6) == "1234")
assert(caesarCipher('', 4) == '')
assert(caesarCipher('zodiac', -2) == 'xmbgya')
assert(caesarCipher('This Is A Difficult Test', -25) == 'Uijt Jt B Ejggjdvmu Uftu')
```
